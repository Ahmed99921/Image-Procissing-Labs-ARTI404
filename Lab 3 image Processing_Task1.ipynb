```python
#Showing the used image.

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('output3.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#Task 1.1:Increase image size
scaled_img = cv2.resize(img, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)

plt.imshow(scaled_img)
plt.title("Scaled Image (1.5x)")
plt.axis("off")
plt.show()
```

```python
#Task 1.2:Rotate the image by 120 degree.
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 120, 1.0)
rotated_img = cv2.warpAffine(img, rotation_matrix, (w, h))

plt.imshow(rotated_img)
plt.title("Task 1.2 - Rotated (120 Degrees)")
plt.axis("off")
plt.show()
```

```python
#Task 1.3:Perform Shear operation.
rows, cols, ch = img.shape

shear_matrix = np.float32([[1, 0.5, 0],
                           [0, 1, 0]])

sheared_img = cv2.warpAffine(img, shear_matrix, (cols + int(cols*0.5), rows))

plt.imshow(sheared_img)
plt.title("Task 1.3 - Shear Transformation")
plt.axis("off")
plt.show()
```
