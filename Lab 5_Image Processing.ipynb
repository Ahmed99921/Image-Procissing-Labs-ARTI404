```python
#Task 1
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('parrot3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

# Apply 7x7 box filter (average blur)
box_filtered = cv2.blur(image, (7, 7))

# Plot original and smoothed images side by side
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(box_filtered)
plt.title("7x7 Box Filter Smoothed")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#Task 2
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('parrot3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply 5x5 Gaussian filter
gaussian_5x5 = cv2.GaussianBlur(image, (5, 5), sigmaX=0, sigmaY=0)

# Apply 21x21 Gaussian filter
gaussian_21x21 = cv2.GaussianBlur(image, (21, 21), sigmaX=0, sigmaY=0)

# Display all three images side by side
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(gaussian_5x5)
plt.title("5x5 Gaussian Filter")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(gaussian_21x21)
plt.title("21x21 Gaussian Filter")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#Task 3
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image
image = cv2.imread('parrot3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 2: Convert image to float32 (range 0 to 1)
image_float = image.astype(np.float32) / 255.0

# Step 3: Apply Laplacian filter
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

# Step 4: Sharpen the image (subtract Laplacian from original)
sharpened = np.clip(image_float - laplacian, 0, 1)

# Step 5: Plot all three images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian Filter Output")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(sharpened)
plt.title("Sharpened Image (Original - Laplacian)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Task 1:Convolve an image with a 7x7 box filter, and output the original and smoothed image side by side
#Read image
image = cv2.imread('parrot3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Create 7x7 box filter kernel
kernel = np.ones((7,7), np.float32) / 49

# Apply convolution
smoothed = cv2.filter2D(image, -1, kernel)

# Plot images
fig, ax = plt.subplots(1, 2, figsize=(10,5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(smoothed)
ax[1].set_title("7x7 Box Filter (Smoothed)")
ax[1].axis("off")

plt.show()
```

```python
#Task 2:Using an image of your choice, apply a 5x5 Gaussian filter to the image, then a 21x21 Gaussian filter, and show the three images side by side.

# Read image
image = cv2.imread('parrot3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply Gaussian filters
gaussian_5 = cv2.GaussianBlur(image, (5,5), 0)
gaussian_21 = cv2.GaussianBlur(image, (21,21), 0)

# Plot images
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(gaussian_5)
ax[1].set_title("5x5 Gaussian")
ax[1].axis("off")

ax[2].imshow(gaussian_21)
ax[2].set_title("21x21 Gaussian")
ax[2].axis("off")

plt.show()
```

```python
#Task 3:Using an image of your choice, apply a 3x3 Laplacian filter to sharpen the image.
# Step 1: Read image
image = cv2.imread('parrot3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 2: Convert to float32
image_float = image.astype(np.float32) / 255.0

# Step 3: Apply Laplacian filter
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

# Step 4: Sharpen image
sharpened = np.clip(image_float - laplacian, 0, 1)

# Step 5: Plot results
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(np.clip(laplacian + 0.5, 0, 1))
ax[1].set_title("Laplacian Image")
ax[1].axis("off")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened Image")
ax[2].axis("off")

plt.show()
```
